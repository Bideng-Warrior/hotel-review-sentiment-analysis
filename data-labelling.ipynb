```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('data/Hotel_Reviews.csv')
```

```python
data
```

```python
data_review = data[['Negative_Review', 'Positive_Review', 'Reviewer_Score']]
data_review
```

```python
data_review['combined_review'] = data_review['Negative_Review'] + data_review['Positive_Review']
```

```python
data_review
```

```python
def label_sentiment(score):
    if score >= 9:
        return 2
    elif score >= 7:
        return 1
    elif score >= 5:
        return 0
    elif score >= 3:
        return -1
    else:
        return -2
    
data_review['sentiment_label'] = data_review['Reviewer_Score'].apply(label_sentiment)
```

```python
data_review
```

Label :
-  2 : Very Positive
-  1 : Positive
-  0 : Neutral
- -1 : Negative
- -2 : Very Negative

```python
final_review = data_review[['combined_review', 'sentiment_label']]
final_review
```

```python
final_review.to_csv('data/review.csv', index=False)
```
